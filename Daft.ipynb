```python
from daftlistings import Daft, SortOrder, SortType, RentType
import pandas as pd
from datetime import date, timedelta
import plotly.express as px
```

```python
today = date.today().strftime('%Y-%m-%d')
today
```

```python
location, price, entered, updated, views, link = [],[],[],[],[],[]
```

```python
%%time

for i in range(1000, 2000, 100):
    print(i, i+99)
    
    daft = Daft()
    
    daft.set_county("Dublin City")
    daft.set_listing_type(RentType.ANY)
    daft.set_sort_order(SortOrder.ASCENDING)
    daft.set_sort_by(SortType.PRICE)
    daft.set_min_price(i)
    daft.set_max_price(i+99)
    
    listings = daft.search()

    for listing in listings:

        location.append(listing.formalised_address)
        print(listing.formalised_address)
        
        price.append(listing.price)
        print(listing.price)
        
        try:
            entered.append(listing.price_change_history[-1]['date'])
            print(listing.price_change_history[-1]['date'])
        except:
            entered.append(None)

        updated.append(listing.date_insert_update)
        print(listing.date_insert_update)

        views.append(listing.views)
        print(listing.views)
        
        link.append(listing.daft_link)
        print(listing.daft_link)

        print("")
```

```python
print(len(location), len(price), len(entered), len(updated), len(views), len(link))
```

```python
# df = pd.DataFrame({'location':location, 'price':price, 'entered':entered, 'updated':updated, 'views':views, 'link':link})
# df = df.drop_duplicates()
# df.to_csv('data/{}.csv'.format(today), drop_index=True)
df= pd.read_csv('data/{}.csv'.format(today))

start_date = date.today() - timedelta(days=7)
end_date = date.today()
df.entered = pd.to_datetime(df.entered).dt.date
df.updated = pd.to_datetime(df.updated).dt.date

df = df[(df.price.notna())
        & ((df.entered>start_date) | (df.entered.isna()))
        & (df.updated>start_date)
#         & (df.entered<end_date)
#         | ((df.updated>start_date) & (df.updated<end_date) & (df.entered.isna())))
#         & (df.price >= 1000))
       ].reset_index(drop=True)

df.views = df.views.astype(int)
df = df.sort_values(by='views', ascending=False).reset_index(drop=True)

print(len(df))

df.head(30).style
```

```python
townlands = pd.read_csv('meta/Dublin Townlands.csv')
townlands = townlands[(townlands.Type=='T') | (townlands.Type=='D')]['Townlands'].to_list()
townlands
```

```python
area = []
for i in range(0, len(df)):
    print (df.location[i])
    found = 0
    values = ['Other']
    for townland in townlands:
        if townland.lower() in df.location[i].lower():
            values.append(townland)
        else:
            pass
    area.append(values[-1])
    print(values)
    print('    '+values[-1])
    print(' ')
```

```python
df['area'] = area
```

```python
viewsbyarea = df.groupby('area')['views'].agg(['count', 'mean', 'std', 'sum']).reset_index()\
                    .sort_values('mean', ascending=True)

pricebyarea = df.groupby('area')['price'].agg(['mean']).reset_index()\
                    .sort_values('mean', ascending=False).rename(columns={'mean':'price'})

data = viewsbyarea.merge(pricebyarea, on='area').reset_index(drop=True)

data.style
```
